```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re 
from bokeh.plotting import figure, show, output_file
from bokeh.models.annotations import Title
from bokeh.io import output_notebook
```

```python
url = 'https://ru.wikipedia.org/wiki/%D0%A2%D0%B5%D0%BF%D0%BB%D0%BE%D0%B2%D0%B0%D1%8F_%D1%8D%D0%BD%D0%B5%D1%80%D0%B3%D0%B5%D1%82%D0%B8%D0%BA%D0%B0_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8#:~:text=%D0%A2%D0%B5%D0%BF%D0%BB%D0%BE%D0%B2%D0%B0%D1%8F%20%D1%8D%D0%BD%D0%B5%D1%80%D0%B3%D0%B5%D1%82%D0%B8%D0%BA%D0%B0%20%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8%20%E2%80%94%20%D0%BE%D1%82%D1%80%D0%B0%D1%81%D0%BB%D1%8C%20%D1%80%D0%BE%D1%81%D1%81%D0%B8%D0%B9%D1%81%D0%BA%D0%BE%D0%B9,%D0%BA%D0%BE%D1%82%D0%B5%D0%BB%D1%8C%D0%BD%D1%8B%D1%85%2C%20%D1%80%D0%B0%D0%B1%D0%BE%D1%82%D0%B0%D1%8E%D1%89%D0%B8%D1%85%20%D0%BD%D0%B0%20%D0%BE%D1%80%D0%B3%D0%B0%D0%BD%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%BE%D0%BC%20%D1%82%D0%BE%D0%BF%D0%BB%D0%B8%D0%B2%D0%B5.'
id_table = 0
df_power = pd.read_html(url, header = 0, index_col = 0)[id_table] 
df_extid = df_power
df_extid
```

```python
# сортировка по имени
df_power = df_power.sort_values(['Название ТЭС'])
print(df_power)
print()
```

```python
df_extid2 = df_extid.copy()
# Создание столбца 'Энергия выработаная  за сутки'
df_extid2['Энергия выработаная за сутки'] = df_extid2['Установленная мощность, МВт'] 
df_extid2['Энергия выработаная за сутки'] = [int(df_extid2['Энергия выработаная за сутки'][i]) * 24 for i in range(1, len(df_extid2['Энергия выработаная за сутки'])+1)]
df_extid2['Установленная мощность, МВт'] = [int(df_extid2['Установленная мощность, МВт'][i]) for i in range(1, len(df_extid2['Установленная мощность, МВт'])+1)]

df_extid2.reset_index(drop=True, inplace=True)

df_Rabota= df_extid2
df_Rabota
```

```python
print('данные из первой строки и второго столбца')
print(df_extid2.iloc[: :, 0 : 2])
print()
```

```python
# 10. Статистические расчеты по сгруппированным данным.
# Суммарная мощность по типа электростанций.
print('Суммарная мощность:',df_extid['Установленная мощность, МВт'].sum())
print()

# Максимальная мощность по типам электростанций.
df_max = df_extid[(df_extid['Установленная мощность, МВт']) == (df_extid['Установленная мощность, МВт'].max())]
print('Максимальная мощность:',df_max[['Название ТЭС', 'Установленная мощность, МВт']])
print()
# Число электростанций по мощности ниже и выше 500 МВт и разделением по компаниям
print(df_extid.groupby(['Собственник', df_extid['Установленная мощность, МВт'] > 2400])['Установленная мощность, МВт'].count())
print()

```

```python
# 13. Гистограмма после группировки
df_extid.groupby(['Название ТЭС'])['Установленная мощность, МВт'].sum().plot(kind = 'barh')
print()
```

```python
# Круговая диаграмма 
df_oblast = df_extid.groupby('Регион').sum()
labels = df_oblast.index
fig1, ax1 = plt.subplots(figsize = (7, 7))
ax1.pie(df_oblast['Установленная мощность, МВт'], labels = df_oblast.index, autopct='%1.1f%%', shadow=True, wedgeprops={'lw':1, 'ls':'--','edgecolor':"k"}, rotatelabels=True)
ax1.axis('equal')  

plt.show()
```
